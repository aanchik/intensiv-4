```python
import pandas as pd
import ast
```

##### После категоризации комментариев получаем файл и читаем его

```python
df1 = pd.read_csv('csvs/Размеченные комментарии.csv')
df1
```

##### Удаляем пустые строки

```python
df1_cleaned = df1.dropna(how='all')
df1_cleaned
```

##### Удаляем не нужные колонки

```python
df1 = df1_cleaned.drop(columns=['annotator', 'annotation_id', 'created_at', 'lead_time', 'updated_at', 'Претензии', 'Предложения'], axis=1)

df1
```

##### Создаем бинарные признаки 1 и 0 для каждой категории, и проверяем содержание в колонке taxonomy, преобразовываем текстовую категоризацию в табличный вид

```python
categories = [
    'Вопрос не решен',
    'Вопрос решен',
    'Не понравилась работа сотрудников',
    'Не понравилась скорость работы',
    'Не понравилось качество услуги',
    'Не понравился результат услуги',
    'Понравилась работа сотрудников',
    'Понравилась скорость работы',
    'Понравилось качество услуги',
    'Понравился результат услуги',
    'Претензии и предложения',
]


for ctgr in categories:
    df1[ctgr] = df1['taxonomy'].apply(
        lambda x: int(
            ctgr in [
                subitem[0]
                for item in ast.literal_eval(x)
                if 'taxonomy' in item
                for subitem in item['taxonomy']
            ]
        ) if x else 0
    )
```

```python
df1
```

```python
df1['Не понравилось качество услуги'] = df1[['Не понравилось качество услуги', 'Не понравилась работа сотрудников']].max(axis=1)
df1['Претензии и предложения'] = df1[['Претензии и предложения', 'Не понравилась скорость работы']].max(axis=1)
df1 = df1.drop(columns=["Не понравилась работа сотрудников", "Не понравилась скорость работы"], axis=1)

df1
```

##### Изменение индекса и удаление колонки taxonomy

```python
df1.reset_index(drop=True, inplace=True)
df1.index += 1

df1 = df1.drop('taxonomy', axis=1)
df1
```

```python
df1 = df1.drop_duplicates(subset="comment")
print(f"Количество уникальных комментариев: {len(df1)}")
```

```python
print(df1['Вопрос не решен'].sum())
print(df1["Вопрос решен"].sum())
print(df1['Не понравилось качество услуги'].sum())
print(df1['Не понравился результат услуги'].sum())
print(df1['Понравилась работа сотрудников'].sum())
print(df1['Понравилась скорость работы'].sum())
print(df1['Понравилось качество услуги'].sum())
print(df1['Понравился результат услуги'].sum())
print(df1['Претензии и предложения'].sum())
```
