##### Создаем бинарные признаки 1 и 0 для каждой категории, и проверяем содержание в колонке taxonomy, преобразовываем текстовую категоризацию в табличный вид

```python
categories = [
    'Вопрос не решен',
    'Вопрос решен',
    'Не понравилась работа сотрудников',
    'Не понравилась скорость работы',
    'Не понравилось качество услуги',
    'Не понравился результат услуги',
    'Понравилась работа сотрудников',
    'Понравилась скорость работы',
    'Понравилось качество услуги',
    'Понравился результат услуги',
    'Претензии и предложения',
]


for ctgr in categories:
    df[ctgr] = df['taxonomy'].apply(
        lambda x: int(
            ctgr in [
                subitem[0]
                for item in ast.literal_eval(x)
                if 'taxonomy' in item
                for subitem in item['taxonomy']
            ]
        ) if x else 0
    )
```

##### Обработка эмодзи, сленговых слов и множественных знаков препинания

```python
# Словарь для перевода часто встречающихся эмодзи на русский
emoji_map_ru = {
    "smiling_face_with_smiling_eyes": "улыбающееся лицо с глазами",
    "thumbs_up": "палец вверх",
    "grinning_face": "широко улыбающееся лицо",
    "face_with_tears_of_joy": "лицо со слезами радости",
    "red_heart": "красное сердце",
    "clapping_hands": "аплодисменты",
    "thinking_face": "задумчивое лицо",
    "winking_face": "подмигивающее лицо",
    "confused_face": "озадаченное лицо",
    "handshake": "рукопожатие",
    "face_with_rolling_eyes": "лицо с закатывающимися глазами",
    "smiling_face_with_hearts": "улыбающееся лицо с сердечками",
    "folded_hands": "сложенные руки (молитва или благодарность)",
    "beaming_face_with_smiling_eyes": "сияющее улыбающееся лицо",
    "smiling_face_with_open_hands": "улыбающееся лицо с открытыми руками",
    "grinning_squinting_face": "широко улыбающееся лицо с прищуренными глазами",
    "raising_hands_light_skin_tone": "поднятые руки (светлая кожа)",
    "cherry_blossom": "вишнёвый цвет",
    "OK_hand": "жест ОК",
    "heart_hands_medium-light_skin_tone": "руки в форме сердца (светлая кожа)",
    "airplane": "самолёт",
    "thumbs_up_light_skin_tone": "палец вверх (светлая кожа)",
    "two_hearts": "два сердца",
    "clapping_hands_light_skin_tone": "аплодисменты (светлая кожа)",
    "smiling_face_with_sunglasses": "улыбающееся лицо в солнцезащитных очках",
    "smiling_face": "улыбающееся лицо",
    "heart_hands": "руки в форме сердца",
    "smiling_cat_with_heart-eyes": "улыбающийся кот с сердечками вместо глаз",
    "blue_heart": "синее сердце",
    "slightly_smiling_face": "слегка улыбающееся лицо",
    "pensive_face": "задумчивое лицо",
    "star-struck": "восхищённое лицо со звёздами в глазах",
    "pink_heart": "розовое сердце",
    "face_holding_back_tears": "лицо сдерживающее слёзы",
    "raising_hands": "поднятые руки",
    "sparkling_heart": "искрящееся сердце",
    "fire": "огонь",
    "thumbs_up_medium_skin_tone": "палец вверх (средний тон кожи)",
    "clapping_hands_medium_skin_tone": "аплодисменты (средний тон кожи)",
    "folded_hands_medium-light_skin_tone": "сложенные руки (светлый средний тон кожи)",
    "Christmas_tree": "рождественская ёлка",
}


def clean_and_convert(text):
    if not isinstance(text, str):
        return ""
     
    # Преобразуем эмодзи в текст
    text = emoji.demojize(text)

    for key, val in emoji_map_ru.items():
        text = text.replace(f":{key}:", " " + val)

    # Переводим известные эмодзи
    for eng_code, ru_text in emoji_map_ru.items():
        text = text.replace(f":{eng_code}:", ru_text)

    return text


# Словарь для замены англ. слов и разговорных выражений
slang_map = {
    "ok": "хорошо",
    "super": "отлично",
    "thanks": "спасибо",
    "pls": "пожалуйста",
    "good": "хорошо",
    "bad": "плохо",
    "sorry": "извините",
    "omg": "о боже",
    "cool": "круто",
    "great": "отлично"
}

# Очистка текста: нижний регистр, замена по словарю, очистка лишнего
def clean_full(text):
    if not isinstance(text, str):
        return ""
    
    text = text.lower()
    text = re.sub(r'_x000d_', ' ', text, flags=re.IGNORECASE)
    text = re.sub(r'[!?.,:;]{2,}', ' ', text)
    text = re.sub(r'[-–—=]{2,}', ' ', text)
    text = re.sub(r'\.{2,}', ' ', text)
    text = re.sub(r'[^а-яa-z0-9\s:]+', '', text)
    text = re.sub(r'\s{2,}', ' ', text)

    # Замены разговорных выражений
    words = text.split()
    words = [slang_map.get(word, word) for word in words]
    return ' '.join(words).strip()

# Применим к колонке
df["comment_clean"] = df["comment"].apply(clean_full)

# Применим к колонке 'comment'
df["comment"] = df["comment"].apply(clean_and_convert)

df = df.drop(columns='comment_clean', axis=1)

```

##### Объединяем бинарные колонки

```python
df['Не понравилось качество услуги'] = df[['Не понравилось качество услуги', 'Не понравилась работа сотрудников']].max(axis=1)
df['Претензии и предложения'] = df[['Претензии и предложения', 'Не понравилась скорость работы']].max(axis=1)
df['Не понравился результат услуги'] = df[['Не понравился результат услуги', 'Не понравилось качество услуги']].max(axis=1)
df = df.drop(columns=["Не понравилась работа сотрудников", "Не понравилась скорость работы", "Не понравилось качество услуги"], axis=1)

df
```
